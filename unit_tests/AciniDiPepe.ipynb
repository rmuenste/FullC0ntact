```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def func_pi(iidx):
    return 15.0 * np.cos(iidx/60.0 * np.pi)

def func_theta(iidx):
    return 15.0 * np.sin(iidx/60.0 * np.pi)
```

```python
n_c = 121
points = np.zeros((n_c,2,3))
for j in range(2):
    for i in range(n_c):
        x = func_pi(i)
        y = func_theta(i)    
        z = float(j)
        points[i,j,0] = x
        points[i,j,1] = y
        points[i,j,2] = z    
```

```python
plt.plot(points[:n_c,0,0],points[:n_c,0,1])
```

```python
plt.scatter(points[:n_c,0,0],points[:n_c,0,1])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(points[:,:,0],points[:,:,1],points[:,:,2])

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
